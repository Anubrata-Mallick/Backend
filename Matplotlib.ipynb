MatPlotLib


*   Usefull for making Plots



```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100) # 100 values between 0 and 10
y = np.sin(x)
z = np.cos(x)
```

**Plotting the data**

```python
# sin wave
plt.plot(x, y)
plt.xlabel('Angle')     # adding x-axis lables
plt.ylabel('Sn value')  # adding y-axis lables
plt.title('sine wave')  # adding title of graph
plt.show()
```

```python
# cosin Wave

plt.plot(x, z)
plt.show()
```

```python

# Parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y) # to plot it using + symbol and red color plot(x, y, 'r+')
plt.show()
```

```python

# multiple line in single graph

x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1]) # 0, 0 is origin and 1, 1 is width and heiht

languages = ['English', 'Latin', 'French', 'German', 'Spanish']
people = [100, 50, 150, 40, 70]

ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('no of People')

plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])

languages = ['English', 'Latin', 'French', 'German', 'Spanish']
people = [100, 50, 150, 40, 70]

ax.pie(people, labels=languages, autopct='%1.1f%%') # autopct => shows the digits & '%1.1f%%' show the decimal format
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])

ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')

plt.show()
```

**3D Scatter Plot**

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Blues')
plt.show()
```
